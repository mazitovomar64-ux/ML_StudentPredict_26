Matplotlib & Seaborn (20 задач)

```python
sns.histplot(data=df, x=df['math score'], bins=20, kde=True, hue='gender')
plt.title('График по math score')
plt.xlabel('баллы по матиматике')
plt.ylabel('количество студентов')
plt.savefig('student grafig.png')
plt.show()
```

```python
#2. Создание простого графика
plt.plot([1, 2, 3], [4, 5, 6], linestyle= '--', marker='o')
plt.title('Линейный plot')
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6], label='Линия 1')
plt.plot([1, 2, 3], [6, 5, 4], label='Линия 2')
plt.legend()
plt.show()
```
